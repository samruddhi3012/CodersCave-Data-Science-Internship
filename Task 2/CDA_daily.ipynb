- Here T = trace precipitation amount or snow depth (an amount too small to measure, usually < 0.005 inches water, so replace it with 0)

```python
cols = ['DailyPrecipitation', 'DailySnowDepth', 'DailySnowfall']

for i in cols:
    if (daily[i] == 'T').any():
        print(f'Column: {i}')
        print(f'Number of values with "T": {(daily[i] == "T").sum()}')
        print("-----------")
```

> #### Anomaly Dectection

```python
# To identify outliers in data
fig, axs = plt.subplots(1, 3, figsize=(15, 5))
columns = ['DailyPrecipitation', 'DailySnowfall', 'DailyAverageRelativeHumidity']

for i, col in enumerate(columns):
    axs[i].boxplot(ppt[col])
    axs[i].set_xlabel(f'{col}')
    axs[i].set_ylabel('mm or inches' if 'Precipitation' in col or 'Snowfall' in col else 'Percentage')
    axs[i].set_title(f'{col}')

plt.tight_layout()
plt.show()
```

- **Due to the high number of trace precipitation or snowfall (T) values being replaced with 0, the actual recorded values now appear as outliers.**

```python
# Skewness present in data
fig, axs = plt.subplots(1, 3, figsize=(15,5))
columns = ['DailyPrecipitation', 'DailySnowfall', 'DailyAverageRelativeHumidity']

for i, col in enumerate(columns):
    sns.distplot(ppt[col], kde=True, ax=axs[i])
    #axs[i].set_xlabel(f'{col}')
    axs[i].set_xlabel('mm or inches' if 'Precipitation' in col or 'Snowfall' in col else 'Percentage')
    axs[i].set_title(f'{col}')

plt.tight_layout()
plt.show()
```

- Skewness is present in precipitation and snowfall data.

> #### Daily Trend

```python
# Daily Trends for Daily Precipitation
years = [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]

fig, ax = plt.subplots(4, 2, figsize=(12, 10))
for i, year in enumerate(years):
    data_for_year = ppt[(ppt['DATE'].dt.year == year)]
    ax[i // 2, i % 2].plot(data_for_year['DATE'], data_for_year['DailyPrecipitation'], color='lightcoral')
    ax[i // 2, i % 2].axhline(data_for_year['DailyPrecipitation'].mean(), color='midnightblue', linestyle='-.')
    ax[i // 2, i % 2].set_xlabel('Dates')
    ax[i // 2, i % 2].set_ylabel('mm or inches')
    ax[i // 2, i % 2].set_title(f'{year}')
    ax[i // 2, i % 2].tick_params(axis='y', labelsize=7)

plt.suptitle('Daily Precipitation',fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# Daily Trend for Snowfall
fig, ax = plt.subplots(4, 2, figsize=(12, 10))
for i, year in enumerate(years):
    data_for_year = ppt[(ppt['DATE'].dt.year == year)]
    ax[i // 2, i % 2].plot(data_for_year['DATE'], data_for_year['DailyAverageRelativeHumidity'], color='lightcoral')
    ax[i // 2, i % 2].axhline(data_for_year['DailyAverageRelativeHumidity'].mean(),color='midnightblue',linestyle='-.')
    ax[i // 2, i % 2].set_xlabel('Dates')
    ax[i // 2, i % 2].set_ylabel('Percent')
    ax[i // 2, i % 2].set_title(f'{year}')
    ax[i // 2, i % 2].tick_params(axis='y', labelsize=7)

plt.suptitle('Daily Average Relative Humidity Plot',fontsize=14)
plt.tight_layout()
plt.show()
```

> #### Yearly Distribution

```python
# Yearly Data
yearly_data = ppt.groupby(ppt['DATE'].dt.year)[['DailyPrecipitation','DailySnowfall','DailyAverageRelativeHumidity']].agg(
    {'DailyPrecipitation':'sum',
     'DailySnowfall':'sum',
     'DailyAverageRelativeHumidity' :'mean'}).reset_index()
```

```python
yearly_data
```

```python
fig,(ax1, ax2) = plt.subplots(1,2,figsize=(11,4))

width = 0.25  
ax1.bar(yearly_data['DATE'] - width/2, yearly_data['DailyPrecipitation'], width, label='Precipitation',color='skyblue')
ax1.bar(yearly_data['DATE'] + width/2, yearly_data['DailySnowfall'], width, label='Snowfall', color='lightcoral')
ax1.set_title('Precipitation and Snowfall')  
ax1.set_xlabel('Years')
ax1.set_ylabel('Total Daily Values')
ax1.legend()

ax2.bar(yearly_data['DATE'], yearly_data['DailyAverageRelativeHumidity'], width = 0.3,color='#e4cbff')
ax2.set_title('Relative Humidity')  
ax2.set_xlabel('Years')
ax2.set_ylabel('Mean Relavtive Humidity (%)')

plt.suptitle('Yearly Distribution', fontsize=14)
plt.tight_layout(pad=1)
plt.show()
```

- Key Points:
1. The trend in snowfall initially showed an upward movement, followed by a subsequent decline after 2018.
2. Maximum Snowfall is observed in year 2018, following 2019 and then 2017.
3. A slight fluctuation is observed in precipitation levels, but the changes are not significant.
4. Maximum Precipitation was observed in 2019, 2021 and 2018.
5. Similarly Maximum Average Relative Humidity was obserd in 2021, 2020 and 2018.
6. From above observation it is clear that during 2018 Albany experinced heavy snowfall and precipitation.

> #### Percentage Changes

```python
# Percentage change from previous year
yearly_data['Precipitation_Change']     = yearly_data['DailyPrecipitation'].pct_change()*100
yearly_data['RelavtiveHumidity_Change'] = yearly_data['DailyAverageRelativeHumidity'].pct_change()*100
yearly_data['Snowfall_Change']          = yearly_data['DailySnowfall'].pct_change()*100
```

```python
yearly_data
```

```python
# Percentage Change from 2015 to 2022

data_2015 = yearly_data[yearly_data['DATE'] == 2015]
data_2022 = yearly_data[yearly_data['DATE'] == 2022]

cal_pre = ((data_2022['DailyPrecipitation'].values - data_2015['DailyPrecipitation'].values) / data_2015['DailyPrecipitation'].values)*100

cal_rh = ((data_2022['DailyAverageRelativeHumidity'].values - data_2015['DailyAverageRelativeHumidity'].values) 
                                                            / data_2015['DailyAverageRelativeHumidity'].values) * 100

cal_snow = ((data_2022['DailySnowfall'].values - data_2015['DailySnowfall'].values) 
                                                     / data_2015['DailySnowfall'].values) * 100
```

```python
print('Changes in precipitation from 2015 to 2022:    ',cal_pre)
print('Changes in relative humidity from 2015 to 2022:',cal_rh)
print('Changes in snowfall from 2015 to 2022:         ',cal_snow)
```

- ### Key Points:
  1. A negative sign indicates a decline in the respective parameter.
  2. Precipitation has decreased by 60.95% from 2015 to 2022.
  3. Relative humidity has decreased by 1.57%.
  4. Snowfall has increased by 34.34% from 2015 to 2022.

> #### Seasonal Variation in temperature

Seasons in New York area :
1. Winter : DJF
2. Spring : MAM
3. Summer : JJA
4. Autumn : SON 

```python
# Create new 'Year' column
temp['Year'] = temp['DATE'].dt.year
```

```python
temp.head()
```

```python
# Create a column season
temp['Season'] = temp['DATE'].dt.month.map({12: 'Winter', 1: 'Winter', 2: 'Winter', 
                                            3: 'Spring', 4: 'Spring', 5: 'Spring',
                                            6: 'Summer', 7: 'Summer', 8: 'Summer', 
                                            9: 'Fall', 10: 'Fall', 11: 'Fall'})
```

```python
temp.head()
```

```python
# Sesonal variation in temperature 

# Adjust the date for plotting
temp['AdjustedDate'] = temp['DATE'].apply(lambda x: x.replace(year=2020) if x.month != 12 else x.replace(year=2019))

unique_years = temp['Year'].unique()
fig, axs = plt.subplots(4, 2, figsize=(15, 15))
axs = axs.flatten()

# Generate list of the first day of each month
month_starts = pd.date_range(start='2019-12-01', end='2020-11-01', freq='MS')

for i, year in enumerate(unique_years):
    data_year = temp[temp['Year'] == year]

    if not data_year.empty:
        sns.lineplot(ax=axs[i], data=data_year, x='AdjustedDate', y='DailyAverageDewPointTemperature', hue='Season')
        axs[i].set_xlabel('Date')
        axs[i].set_ylabel('(Â°C)')
        axs[i].set_title(f'Year {year}')
        axs[i].legend(loc="upper left")
        
        # Set x-ticks to the start of each month and format labels to show month names
        axs[i].set_xticks(month_starts)
        axs[i].set_xticklabels(month_starts.strftime('%b'))

plt.suptitle('Seasonal Variation of Daily Average Dewpoint Temperature', fontsize=15)
plt.tight_layout(pad=2)
plt.show()
```

- The seasonal temperature trend is represented above.
